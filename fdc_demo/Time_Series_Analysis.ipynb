```python
final_data['mavg_12']= final_data['TRx_Derived'].rolling(window = 12).mean().shift(1)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
final_data[['TRx_Derived', 'mavg_12']][12:]
```

```python
plt.figure(figsize=(10,4))
plt.xlabel("Months")
plt.ylabel("TRx")
plt.plot(final_data['TRx_Derived'][12:])
plt.plot(final_data['mavg_12'][12:], '.')
plt.legend()
```

```python
#Defining Mape Function
import numpy as np
def get_mape(actual, predicted):
    y_true , y_pred = np.array(actual) , np.array(predicted)
    return np.round(np.mean(np.abs((actual-predicted)/actual))*100, 2)
```

```python
get_mape(final_data['TRx_Derived'][12:].values , final_data['mavg_12'][12:].values)
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(final_data['TRx_Derived'][12:].values , final_data['mavg_12'][12:].values))
```

```python
#Exponential Smoothing
final_data['ewm'] = final_data['TRx_Derived'].ewm(alpha = 0.2).mean()
final_data[12:]
```

```python
mape_error = get_mape(final_data['TRx_Derived'][12:].values , final_data['ewm'][12:].values)
mean_squared_error_value = np.sqrt(mean_squared_error(final_data['TRx_Derived'][12:].values , final_data['ewm'][12:].values))
print('mape = ', mape_error)
print('mean_squared_error = ', mean_squared_error_value )
```
